```python
import datetime as dt
```

```python

menu={
    "Burger": 150,
    "Pizza": 250,
    "Pasta": 200,
    "Salad": 120,
    "Beverages": 130,
    "Noodles": 150,
    "Sushi": 270,
    "Bakery":350,
            }
    
orderID=0
Detail=[]    
counter=0 
while counter == len(Detail):
    order =str(input('Digitize food:'))
    if order in menu:
        Detail.append(order)
        counter +=1
    elif order == 'v':
        verificacion=set(Detail)
        for f in verificacion:            
            print('{:^10}{:^10}{:^10}'.format(f,'',Detail.count(f)))
        print('*****************')
    elif order=='a':
        add= str(input('Add a Food:'))
        if add in menu:
            Detail.append(add)
            counter+=1
            order
        else: 
            print(f'{add} in not in menu')
            order
    elif order=='r':
        delete=str(input('Delete a foot:'))
        if delete in Detail:
            Detail.remove(delete)
            counter-=1
            order
        else:
            print(f'{delete} in not in the order')
            order 
    elif order =='x' and len(Detail)>=1:     
        orderID +=1
        print('Complete order')
        print (Detail)
        print(orderID)

        break
    else:
        print(f'{order} is not available in menu')
    

total_bill=0
for ord in Detail:
    if ord in menu:
        total_bill+= menu[ord]
        print('{:^10}{:^10}{:^10}'.format(f'{ord}','',f'{menu[ord]}'))
print('*******************')
ahora=dt.datetime.now()
print(ahora)
final=set(Detail)
print('{:^10},{:^10},{:^10}{:^10}'.format('Description','Amount','U. price','Subtotal'))
for dishe in final:    
    print('{:^10},{:^10},{:^10}{:^10}'.format(f'{dishe}',f'{Detail.count(dishe)}',f'{menu[dishe]}',f'{Detail.count(dishe)*menu[dishe]}'))
print('{:^10},{:^10},{:^10}{:^10}'.format('****','****','****','****'))
if total_bill>=1000:
    print('{:^10}{:^10}{:^10}{:^10}'.format('ITEMS','','',f'{counter}'))
    print('{:^10}{:^10}{:^10}{:^10}'.format('SUBTOTAL','','',f'{total_bill}'))
    print('{:^10}{:^10}{:^10}{:^10.2f}'.format('Service','','',(total_bill*0.1)))
    print('{:^10}{:^10}{:^10}{:^10.2f}'.format('VAT','','',(total_bill*0.16)))
    print('{:^10}{:^10}{:^10}{:^10.2f}'.format('TOTAL','','',(total_bill*1.26)))
else:
    print('{:^10}{:^10}{:^10}{:^10}'.format('ITEMS','','',f'{counter}'))
    print('{:^10}{:^10}{:^10}{:^10}'.format('TOTAL','','',F'{total_bill}'))


File= open('Sales.txt','a')
File.write(f'{str(ahora)},{str(counter)},{str(total_bill)},{str(Detail)}.\n')
File.close()
```
